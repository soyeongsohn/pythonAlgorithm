# Graph
- 사물을 정점(vertex)과 간선(edge)으로 나타내기 위한 도구
- 두 가지 방식으로 구현 가능:
1. 인접 행렬(adjacency matrix): 2차원 배열을 사용하는 방식 - 각각의 노드가 연결되어있는지 여부를 배열로 표현
2. 인접 리스트(adjacency list): 연결 리스트를 사용하는 방식 - 각각 노드마다 하나의 리스트를 만들어서 그 노드가 어떤 노드와 연결되어 있는지 확인

## 인접 행렬 (Adjacency Matrix)

### 무방향 무가중치 그래프
- __무방향 그래프__: 모든 간선이 방향성을 가지지 않는 그래프
- __무가중치 그래프__: 모든 간선에 가중치가 없는 그래프


### 방향 가중치 그래프
- __방향 그래프__: 모든 간선이 방향을 가지는 그래프
- __가중치 그래프__: 모든 간선에 가중치가 있는 그래프
